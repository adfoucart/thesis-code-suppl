Generation of the graph comparing the activation functions:
* Logistic function
* Hyperbolic tangent
* ReLU
* Leaky ReLU

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def logistic_function(x):
    return (1/(1+np.exp(-x)))

def relu(x):
    y = x.copy()
    y[x<0] = 0
    return y

def leaky_relu(x, l=0.01):
    y = x.copy()
    y[x<0] = l*x[x<0]
    return y

x = np.arange(-5, 5, 0.01)

plt.figure(figsize=(10,7))
plt.plot(x, logistic_function(x), label='logistic')
plt.plot(x, np.tanh(x), label='tanh')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, leaky_relu(x, 0.1), label='Leaky ReLU')
plt.ylim([-1.5, 1.5])
plt.xlabel('x')
plt.ylabel('a(x)')
plt.title('Activation functions')
plt.legend()
plt.show()
```
