# Chapter 2 - Digital pathology

Computing the difference in hue histograms between and H&E-stained histopathological image and a natural photographic image.

* H&E-stained image comes from [MoNuSAC data](monusac-2020.grand-challenge.org/data/)
* Photograph by Adrien Foucart.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from skimage.io import imread
from skimage.color import rgb2hsv, hsv2rgb

def plot_hue(im: np.array, show: bool = True) -> None:
    hsv = rgb2hsv(im[..., :3])
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(im)
    plt.subplot(1, 2, 2)
    n, bins, patches = plt.hist(hsv[..., 0].flatten(), bins=360)
    for i,p in enumerate(patches):
        bin_center = (bins[i]+bins[i+1])/2
        plt.setp(p, 'facecolor', hsv2rgb([bin_center, 1, 1]))
    plt.xlabel('Hue')
    plt.ylabel('N pixels')
    plt.show()
```

```python
monusag_im = imread("./data/monusac_image.tif")
etretat_im = imread("./data/etretat.jpg")
```

```python
plot_hue(monusag_im, show=False)
plot_hue(etretat_im)
```
